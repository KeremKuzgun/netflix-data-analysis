# Netflix Data Analysis

This project analyzes Netflix content data using Python.

## Technologies
- Python
- Pandas
- NumPy
- Matplotlib

##  Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

## Explore Dataset

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

## Missing Data Analysis

```python
df.isnull().sum()
```

## Data Cleaning

```python
df['country'] = df['country'].fillna("Unknown")
```

## Movie vs TV Show Distribution

```python
df['type'].value_counts()
```

```python
df['type'].value_counts().plot(kind='bar')

plt.title("Netflix Content Types")
plt.xlabel("Type")
plt.ylabel("Count")

plt.show()
```

## Content Released by Year

```python
df['release_year'].value_counts().sort_index().plot()

plt.title("Content by Year")
plt.xlabel("Year")
plt.ylabel("Number of Shows")

plt.show()
```

## Top Content Producing Countries

```python
df['country'].value_counts().head(10)
```

```python
df['country'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")

plt.show()
```

## Rating Distribution

```python
df['rating'].value_counts().plot(kind='bar')

plt.title("Content Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.show()
```

## Statistical Analysis

```python
years = df['release_year']

print("Average Release Year:", np.mean(years))
print("Median Release Year:", np.median(years))

print("Oldest Content:", np.min(years))
print("Newest Content:", np.max(years))

print("Standard Deviation:", np.std(years))

print("Total Titles:", len(years))

print("Most Common Release Year:", years.mode()[0])
```

```python
plt.boxplot(years)

plt.title("Release Year Distribution")

plt.show()
```

## Conclusion

- Movies are more common than TV shows on Netflix.
- Content production increased significantly after 2015.
- The United States produces the most Netflix content.
- TV-MA is the most common content rating.
